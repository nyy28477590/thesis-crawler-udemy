```python
url = 'https://www.udemy.com/api-2.0/discovery-units/all_courses/'
payload = {'p':1,'page_size':60, 'lang':'en', 'category_id':288, 'source_page':'category_page', 'locale':'en_US', 'currency':'usd', 'navigation_locale':'en_US', 'sos':'pc','fl':'cat'}
r = requests.get(url, params=payload)
if r.status_code == 200:
    api_data = r.json()
```

```python
# remove 288
categories = [267,268,269,270,271,272,273,274,275,276,277,278,279,280,281,282,283,290,292,294,296,298,300,302,308,310,312,314,322,324,326,330,332]
```

```python
for c in categories:
    #
    path = './backup/cat-'+str(c)+'/'
    if not os.path.isdir(path):
        os.makedirs(path)
    
    print("Category: ", c)
    courses_info = []
    intructors_info = []

    remaining_items = 1
    page_num = 1
    while (remaining_items!=0):
        print("page: ", page_num)
        print("Remaining page: ", remaining_items)
        url = 'https://www.udemy.com/api-2.0/discovery-units/all_courses/'
        payload = {'p':page_num,'page_size':60, 'lang':'en', 'category_id':c, 'source_page':'category_page', 'locale':'en_US', 'currency':'usd', 'navigation_locale':'en_US', 'sos':'pc','fl':'cat'}
        r = requests.get(url, params=payload)
        if r.status_code == 200:
            api_data = r.json()
            #Remaing_item
            remaining_items = api_data['unit']['remaining_item_count']

            #
            for data in api_data['unit']['items']:
                courses_temp_list = []

                course_id = data['id']
                course_name = data['title']
                course_url = data['url']
                students_count = data['num_subscribers']
                rating = data['rating']
                lectures_count = data['num_published_lectures']
                practice_count = data['num_published_practice_tests']
                course_leve = data['instructional_level']
                published_time = data['published_time']
                updated_time = data['last_update_date']
                course_hours = data['content_info']

                courses_temp_list.append(course_id)
                courses_temp_list.append(course_name)
                courses_temp_list.append(course_url)
                courses_temp_list.append(students_count)
                courses_temp_list.append(rating)
                courses_temp_list.append(lectures_count)
                courses_temp_list.append(practice_count)
                courses_temp_list.append(course_leve)
                courses_temp_list.append(published_time)
                courses_temp_list.append(updated_time)
                courses_temp_list.append(course_hours)

                courses_info.append(courses_temp_list)

                #
                for instructor in data['visible_instructors']:
                    instructor_temp_list = []

                    instructor_id = instructor['id']
                    instructor_name = instructor['title']
                    instructor_job_title = instructor['job_title']
                    instructor_url = instructor['url']

                    instructor_temp_list.append(course_id)
                    instructor_temp_list.append(instructor_id)
                    instructor_temp_list.append(instructor_name)
                    instructor_temp_list.append(instructor_job_title)
                    instructor_temp_list.append(instructor_url)

                    intructors_info.append(instructor_temp_list)

            #json
            with open('./backup/cat-'+str(c)+'/'+'cat-'+str(c)+'-'+'page-'+str(page_num)+'.json', 'w') as fp:
                json.dump(api_data, fp)        
        #
        page_num += 1
        
    #Category,
    try:
        courses_df = pd.DataFrame(courses_info)
        courses_df.columns =['course_id', 'course_name', 'course_url', 'students_count', 'rating', 'lectures_count', 'practice_count', 'course_leve', 'published_time', 'updated_time', 'course_hours']
        courses_df.to_excel('./backup/cat-'+str(c)+'/courses_info_cat_'+str(c)+'.xlsx')

        instructors_df = pd.DataFrame(intructors_info)
        instructors_df.columns =['course_id', 'instructor_id', 'instructor_name', 'instructor_job_title', 'instructor_url']
        instructors_df.to_excel('./backup/cat-'+str(c)+'/instructors_info_cat_'+str(c)+'.xlsx')
    except:
        print('No Data')
        continue
```
